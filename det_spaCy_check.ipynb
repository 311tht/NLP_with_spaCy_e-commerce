```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
my_text = "Looking for a pair of durable running shoes with good arch support on sale."
doc = nlp(my_text)
for token in doc:
    #print(f'token: {token}, day la .dep_: {token.dep_} v√† pos_: {token.pos_}')
    #print(list(token.subtree), '--', token.dep_)
    if ("dobj" in token.dep_):
        subtree = list(token.subtree)
        start = subtree[0].i
        end = subtree[-1].i + 1
        extract = doc[start:end]
        product = token
        print(token.text)
        print(f'RESULT DOBJ: {doc[start:end]}')
    elif ("pobj" in token.dep_):
        subtree = list(token.subtree)
        start = subtree[0].i
        end = subtree[-1].i + 1
        extract = doc[start:end]
        product = token
        print(token.text)
        print(f'RESULT POBJ: {doc[start:end]}')
        print()

```

```python
def get_subject_phrase(doc):
    for token in doc:
            subtree = list(token.subtree)
            start = subtree[0].i
            end = subtree[-1].i + 1
            return doc[start:end]
```
